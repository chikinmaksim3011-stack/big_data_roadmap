```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
df = pd.read_csv('netflix_titles.csv')
result = (df['type'].value_counts())
plt.figure(figsize = (8,8))
sns.barplot(x = result.values, y = result.index, data = result)
plt.title = 'How much types in dataset'
plt.xlabel = 'Names'
plt.ylabel = 'How much'
plt.show()
```

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
df = pd.read_csv('netflix_titles.csv')
df['main_country'] = df['country'].str.split(',').str[0]
result = (df['main_country'].value_counts()).head(10)
plt.figure(figsize = (8,8))
sns.barplot(x = result.values, y = result.index, data = result, orient = 'h')
plt.title = 'Top 10 countries in IMDB'
plt.xlabel = 'Countries'
plt.ylabel = 'Counts'
plt.tight_layout()
plt.show()
```
