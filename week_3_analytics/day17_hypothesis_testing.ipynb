```python
# We are sure, that 2010s movies are more than 2000s because a lot of new technical systems creating movies, ChatGPT and remake old movies create new movies

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../Our Datasets/imdb_top_1000.csv')
df['Released_Year']= pd.to_numeric(df['Released_Year'], errors='coerce')
df['Runtime'] = pd.to_numeric(df['Runtime'].str.split(' ').str[0], errors='coerce')

df = df.dropna(subset=['Released_Year', 'Runtime'])
df_2000s = df[(df['Released_Year'] >= 2000) & (df['Released_Year'] <= 2009)]
avg_time_2000s = df_2000s['Runtime'].mean()
avg_count_2000s = df_2000s['Runtime'].count()
print(avg_time_2000s, avg_count_2000s)

df_2010s = df[(df['Released_Year'] >= 2010) & (df['Released_Year'] <= 2020)]
avg_time_2010s = df_2010s['Runtime'].mean()
avg_count_2010s = df_2010s['Runtime'].count()

print(avg_time_2010s, avg_count_2010s)
```
