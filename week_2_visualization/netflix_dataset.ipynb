```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../Our Datasets/netflix_titles.csv')
result = (df['type'].value_counts())
plt.figure(figsize = (10,6))
sns.barplot(x = result.index, y = result.values, orient = 'v')
plt.title('How much types in dataset')
plt.xlabel('Names')
plt.ylabel('How much')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../Our Datasets/netflix_titles.csv')
df['main_country'] = df['country'].str.split(',').str[0]
result = (df['main_country'].value_counts()).head(10)
plt.figure(figsize = (8,8))
sns.barplot(x = result.values, y = result.index, orient = 'h')
plt.title('How many types in dataset')
plt.xlabel('Count')
plt.ylabel('Type')
plt.tight_layout()
plt.show()
```
