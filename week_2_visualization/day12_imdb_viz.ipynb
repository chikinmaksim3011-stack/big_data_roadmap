```python
# Importing libraries for working with database
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../Our Datasets/imdb_top_1000.csv') # Importing our database in our IDE.
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce') # Converting one of the column to integer for comforting
df.dropna(subset = ['Released_Year'], inplace=True) # Delete None objects in our column (subset - our column/s, inplace - in our database, not creating new)

plt.figure(figsize = (10,10)) # Creating new table for create our diagram (figsize - our size, x and y)
plt.hist(x = df['IMDB_Rating'], bins=20, color='lightcoral', edgecolor='black') # Creating histogram for our database
plt.title('IMDB Rating for Top 1000 Movies')
plt.xlabel('IMDB Rating')
plt.ylabel('Number of Movies')
plt.savefig('imdb_rating_hist.png', dpi = 150, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Our Datasets/imdb_top_1000.csv')
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df.dropna(subset = ['Released_Year'], inplace=True)
df['Genres'] = df['Genre'].str.split(',').str[0]
result = df['Genres'].value_counts().head(5)

plt.figure(figsize = (10,10))
sns.barplot(x = result.values, y = result.index, orient = 'h')
plt.title('Top 5 Genres in Top 1000 Movies')
plt.xlabel('How many Genres were released')
plt.ylabel('Genres')
plt.savefig('imdb_top_genres.png', dpi = 150, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../Our Datasets/imdb_top_1000.csv')
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df.dropna(subset = ['Released_Year'], inplace=True)

plt.figure(figsize = (10,10))
plt.scatter(df['Released_Year'], df['IMDB_Rating'], alpha=0.6, color='teal')
plt.title('Release Year vs. IMDB Rating')
plt.xlabel('Released Year')
plt.ylabel('IMDB Rating')
plt.savefig('imdb_year_vs_rating.png', dpi = 150, bbox_inches='tight')
plt.show()
```
